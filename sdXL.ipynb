```python
from diffusers import AutoPipelineForText2Image
import torch
import matplotlib.pyplot as plt

pipeline = AutoPipelineForText2Image.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16, variant="fp16"
).to("cuda")
generator = torch.Generator("cuda").manual_seed(31)
image = pipeline("anime girl, T-pose,pink hair, detailed, 8k", generator=generator).images[0]

# Convert the image tensor to a numpy array and display it
plt.imshow(image)
plt.axis('off')  # Hide axis
plt.show()
```

```python
from diffusers import AutoPipelineForText2Image
import torch
import matplotlib.pyplot as plt

pipeline = AutoPipelineForText2Image.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, variant="fp16"
).to("cuda")
generator = torch.Generator("cuda").manual_seed(30)
image2 = pipeline("anime girl, T-pose", generator=generator).images[0]

# Display and save the image
plt.imshow(image2)
plt.axis('off')  # Hide axis
plt.savefig('/data/anime_girl_T_pose.jpg', bbox_inches='tight', pad_inches=0)
plt.show()
```
